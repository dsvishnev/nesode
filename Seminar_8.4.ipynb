Решим уравнение $$\ddot{x}=x^2-1.$$

Для этого перепишем его в виде системы

$$\begin{cases}
        \dot x=y\\\\
        \dot y=x^2-1
  \end{cases}$$

и решим ее 
    $$\frac{dy}{dx} = \frac{x^2-1}{y},$$
    
   $$\frac{y^2}{2} - \int (x^2-1) dx = C,$$
   
   $$\frac{y^2}{2} - \frac{x^3}{3} + x = C.$$
   
   Поскольку левая часть равенства выше является первым интегралом, само равенство задает фазовые кривые, которые мы и хотим нарисовать.
        
   Тогда $$y = \pm \sqrt{\frac{2x^3}{3}-2x+C}.$$
   
   Чтобы нарисовать эскиз, рассмотрим по очереди несколько значений $С$, изобразим для них отдельно функцию под корнем и затем нарисуем график корня, зная, как ведет себя такое преобразование. Получим

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

plt.figure(figsize=(6,6))
plt.xlabel("$x$", fontsize=20)
plt.ylabel("$y$", fontsize=20)

y, x = np.mgrid[-4:4:26j, -4:4:26j]

plt.streamplot(x, y, y, x**2 - 1)
```
