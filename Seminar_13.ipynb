## Задача 1 (а)

Найти решение задачи Коши $(x(0),y(0))=(x_0,y_0)$, построить фазовый портрет и определить тип особой точки $(0,0)$.

$$
\begin{cases}
\dot x=- 4 x - 3 y\\
\dot y=6 x + 5 y.
\end{cases}
$$

### Решение:

Перепишем систему в матричном виде

$$\begin{pmatrix}
  \dot{x}\\
  \dot{y}
 \end{pmatrix}
 =
 \begin{pmatrix}
  -4 & -3 \\
  6 & 5
 \end{pmatrix}
 \begin{pmatrix}
  x\\
  y
 \end{pmatrix}
 $$

и диагонализуем матрицу
 $$\begin{pmatrix}
  -4 & -3 \\
  6 & 5
 \end{pmatrix}.$$

Для этого найдем ее собственные значения и векторы

 $$\begin{vmatrix}
  -4-\lambda & -3 \\
  6 & 5-\lambda
 \end{vmatrix}
= 
(-4-\lambda)(5-\lambda)+18
=
\lambda^2-\lambda-2=(\lambda+1)(\lambda-2).$$
То есть $\lambda_1=-1, \; \lambda_2=2$. Значит, система имеет особую точку типа седло.

Найдем $v_1$
$$
\begin{pmatrix}
  -4 & -3 \\
  6 & 5
\end{pmatrix}
\begin{pmatrix}
  x_1 \\
  x_2
\end{pmatrix}
=
\begin{pmatrix}
  -x_1 \\
  -x_2
\end{pmatrix},
$$
откуда $-x_1=x_2$ и $v_1=(-1, 1).$

Найдем $v_2$
$$
\begin{pmatrix}
  -4 & -3 \\
  6 & 5
\end{pmatrix}
\begin{pmatrix}
  x_1 \\
  x_2
\end{pmatrix}
=
\begin{pmatrix}
  2x_1 \\
  2x_2
\end{pmatrix},
$$
откуда $-2x_1=x_2$ и $v_1=(-1, 2).$

Тогда 
$$
\begin{pmatrix}
  -4 & -3 \\
  6 & 5
\end{pmatrix}
=
\begin{pmatrix}
  -1 & -1 \\
  1 & 2
\end{pmatrix}
\begin{pmatrix}
  -4 & -3 \\
  6 & 5
\end{pmatrix}
\begin{pmatrix}
  -1 & -1 \\
  1 & 2
\end{pmatrix}^{-1}
$$
или
$$
\begin{pmatrix}
  -4 & -3 \\
  6 & 5
\end{pmatrix}
=
\begin{pmatrix}
  -1 & -1 \\
  1 & 2
\end{pmatrix}
\begin{pmatrix}
  -4 & -3 \\
  6 & 5
\end{pmatrix}
\begin{pmatrix}
  -2 & -1 \\
  1 & 1
\end{pmatrix}.
$$

Таким образом,
$$
 \begin{pmatrix}
  x\\
  y
 \end{pmatrix}
 =
 \begin{pmatrix}
  -1 & -1 \\
  1 & 2
\end{pmatrix}
\begin{pmatrix}
  e^{-t} & 0 \\
  0 & e^{2t}
\end{pmatrix}
\begin{pmatrix}
  -2 & -1 \\
  1 & 1
\end{pmatrix}
 \begin{pmatrix}
  x_0\\
  y_0
 \end{pmatrix}.
$$

Наконец, построим фазовый портрет.

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

plt.figure(figsize=(6,6))
plt.xlabel("$x$", fontsize=20)
plt.ylabel("$y$", fontsize=20)

y, x = np.mgrid[-8:8:26j, -8:8:26j]

plt.streamplot(x, y, -4*x-3*y, 6*x+5*y)
```

## Задача 2

    $$\dot x=2x+y,\quad \dot y=2y.$$

a) Решить уравнение на $y$.


b) Подставить полученное решение в уравнение на $x$. Решить получающееся уравнение на $x$.


c) Записать решение задачи Коши с начальным условием $(x(0),y(0))=(x_0,y_0)$ в виде


$$\begin{pmatrix}
    x(t)\\
    y(t)
\end{pmatrix}=
M(t)
\begin{pmatrix}
    x_0\\
    y_0
\end{pmatrix},
$$
где $M(t)$ — некоторая матрица.

### Решение:

a) Из курса мы знаем, что $y=y_0e^{2t}$. 

b) Тогда $\dot{x}=2x+y_0e^{2t}.$ Решение этого уравнения (можно найти, например, методом вариации постоянной) $x=y_0te^{2t}+x_0e^{2t}.$

с) Таким образом,
$$\begin{pmatrix}
    x(t)\\
    y(t)
\end{pmatrix}=
 \begin{pmatrix}
  e^{2t} & te^{2t} \\
  0 & e^{2t}
\end{pmatrix}
\begin{pmatrix}
    x_0\\
    y_0
\end{pmatrix}.
$$
